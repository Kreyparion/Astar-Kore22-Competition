```python
from lib import make
from main import *
import json
def agent_test(obs: Observation, config: Configuration):
    if obs.step < 200:
        return agent(obs,config)
    else:
        return dict()
```

```python

kore_map = [[0 for i in range(21)] for j in range(21)]
kore_map[3][3] = 100
dictionary = dict()
dictionary["kore_map"] = kore_map
dictionary["nb_ships"] = 100
dictionary["nb_steps"] = 50
env = make("kore_fleets", debug=True)
print(env.name, env.version)
with open("lib/envs/kore_fleets/config_map.json", "w") as outfile:
    json.dump(dictionary, outfile)
env.run([agent_test])
env.render(mode="ipython", width=1000, height=800)
```
